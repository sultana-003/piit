### LEFT JOIN:

### RIGHT JOIN:

```python
pd.read_sql('select * from customer2 RIGHT JOIN payment ON customer2.customer_id = payment.customer_id;', engine1)
```

### FULL JOIN/ UNION :

```python
pd.read_sql('select * from customer2 LEFT JOIN payment ON customer2.customer_id = payment.customer_id UNION select* from customer2 RIGHT JOIN payment ON customer2.customer_id = payment.customer_id;', engine1)
```
