```python
pd.read_sql('select department, count(salary) as total from employee group by department;', engine1)
```

```python
pd.read_sql('select department, sum(salary) as total from employee group by department;', engine1)
```

```python
pd.read_sql('select department, avg(salary) as total from employee group by department;', engine1)
```

```python
pd.read_sql('select first_name, sum(salary) as total_pay from employee group by first_name having sum(salary)>=100000 order by total_pay;', engine1)
```

```python
pd.read_sql('select * from payment;', engine1)
```

```python
pd.read_sql('select mode, sum(amount) as total from payment group by mode having sum(amount) order by total;', engine1)
```
