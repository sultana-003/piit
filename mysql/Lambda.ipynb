### Run SQL Commands importing from pydataset using Lambda Function:

```python
import pandas as pd
from pandasql import sqldf
import os
from pydataset import data
```

```python
mtcars = data('mtcars')
```

```python
pysqldf = lambda q: sqldf(q, globals())
```

```python
pysqldf("select * from mtcars where mpg > 25;")
```

```python
df = pysqldf("select * from mtcars order by hp desc;")
df.head()
```
