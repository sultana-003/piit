### Task 3

##### Missing value

```python
df.isnull().sum().sum()    # Total missing value
```

```python
df.isnull().sum(axis=0)   # missing value columns 
```

```python
df_mv = df.copy()
```

```python
df_mv['Ozone'] = df_mv['Ozone'].fillna(df_mv['Ozone'].median())   # 'Ozone' fill with median
df_mv.head()
```

```python
df_mv['Solar.R'] = df_mv['Solar.R'].fillna(df_mv['Solar.R'].mean())    # 'Solar.R' fill with mean
df_mv.head()
```

```python
## Briefly comment why might median be preferred over mean.

   #  Median represent the middle value so it gives a better central tendency.
    # Mean deals with outliers. Which can be much higher or lower than the rest
    # of the data making it less representative. The median better effects the 
    # typical value than mean.
```
