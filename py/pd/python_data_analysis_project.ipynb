```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\analytics\data\exam_scores.csv")
df.head()
```

### change weekly study hours column

```python
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace('10-May','5-10')
```

```python
df.head(2)
```

### drop unnamed column

```python
df = df.drop(['Unnamed: 0'], axis=1)
```

```python
df.shape[1]
```

```python
df.columns
```

### gender distribution

```python
ax = sns.countplot(data = df, x = 'Gender')
plt.figure(figsize = (3,1))
ax.bar_label(ax.containers[0])
plt.show()
```

#### from the above chart we have analyzed that the number of females in the data is more than the number of males.

```python
gp = df.groupby('ParentEduc')[['MathScore','ReadingScore','WritingScore']].mean()
```

```python
gp
```

```python
sns.heatmap(gp, annot=True)
plt.title("relationship between parent's education and student's scores")
plt.show()
```

#### from the above chart we have concluded that the education of the parents have a good impact on the students score.

```python
gp1 = df.groupby('ParentMaritalStatus')[['MathScore','ReadingScore','WritingScore']].mean()
gp1
```

```python
sns.heatmap(gp1, annot=True)
plt.title("relationship between Parents Marital Status and student's scores")
plt.show
```

#### from the above chart we have concluded that there is no/negligible impact on the students scores due to their parents marital status.
