```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [50,60,52,55,62]
plt.plot(x,y,color='red',linewidth=5,linestyle='dotted')
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather Chart')
```

```python
x1= np.arange(1,11)
x1
```

```python
y1= 2*x1
y1
```

```python
plt.plot(x1,y1,color='g',linewidth=4,linestyle='dashed')
plt.xlabel('weight',fontsize=15)
plt.ylabel('height')
plt.title('Student')
plt.show()
```

```python
student= {'Bob':78,'Matt':56,'Sam':34}
names=list(student.keys())
values=list(student.values())
plt.bar(names,values,color='g')
plt.title('Bar Plot')
plt.xlabel('Names')
plt.ylabel('Age')
plt.grid(True)
plt.show()
```
