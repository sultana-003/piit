```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
df.shape
```

```python
df.info()
```

### Scatter Plot :

```python
sns.scatterplot(data=df, x='age', y='fare', hue='pclass', palette='bright', size='pclass')
plt.title('relationship between age and fare')
plt.show()
```

### Histogram :

```python
sns.histplot(data=df, x='age', bins=15, kde=True, hue='pclass')
plt.title('age count by pcalss')
plt.show()
```

### Barplot :


```python
sns.barplot(data=df, x='sex', y='fare', palette='deep', hue='pclass')
plt.title('male/female fare')
plt.show()
```

### Boxplot :

```python
sns.boxplot(data=df, x='pclass', y='fare', hue='sex', palette='cool')
plt.title('showing fare by pclass')
plt.show()
```

### Violin Plot :

```python
sns.violinplot(data=df, x='pclass', y='fare', hue='sex', palette='coolwarm')
plt.title('showing fare by pclass')
plt.show()
```
